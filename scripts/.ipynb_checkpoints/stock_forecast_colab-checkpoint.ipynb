# Stock Prediction Model - StokManager
## Linear Regression + Gradient Boosting Regressor

**Project:** StokManager (webinvesp32)
**Referensi:** [Inventory-Warehouse_Stock_Prediction](https://github.com/parita2003/Inventory-Warehouse_Stock_Prediction)

Notebook ini menguji model prediksi stok menggunakan:
1. **Linear Regression** (baseline, interpretable)
2. **Gradient Boosting Regressor** (advanced, higher accuracy)

**Fitur:**
- Dummy data 8 produk + 60 hari history transaksi realistis
- Feature engineering: day_of_week, is_weekend, rolling average
- One-hot encoding kategori produk
- MinMaxScaler normalisasi
- Train/test split kronologis (80/20)
- Metrik: MAE, RMSE, R², Accuracy-like (threshold 5%)
- Forecast 14 hari + estimasi stockout
- Visualisasi lengkap: scatter, residuals, Q-Q plot, training loss

## 1. Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print('All libraries loaded successfully')
```

## 2. Dummy Data - Produk & History Transaksi

```python
# Daftar produk toko
products = [
    {"id": "P001", "name": "76 Apel", "barcode": "8991906106250", "category": "Rokok", "price": 15000, "minStock": 5, "capacity": 50, "currentQty": 23},
    {"id": "P002", "name": "Indomie Goreng", "barcode": "8996001600016", "category": "Makanan", "price": 3500, "minStock": 10, "capacity": 100, "currentQty": 45},
    {"id": "P003", "name": "Aqua 600ml", "barcode": "8886008101053", "category": "Minuman", "price": 4000, "minStock": 15, "capacity": 80, "currentQty": 12},
    {"id": "P004", "name": "Teh Pucuk 350ml", "barcode": "8996001600108", "category": "Minuman", "price": 4500, "minStock": 10, "capacity": 60, "currentQty": 38},
    {"id": "P005", "name": "Roti Sari Roti Coklat", "barcode": "8992907777010", "category": "Makanan", "price": 7500, "minStock": 8, "capacity": 40, "currentQty": 5},
    {"id": "P006", "name": "Sabun Lifebuoy", "barcode": "8999999527273", "category": "Toiletries", "price": 3000, "minStock": 5, "capacity": 50, "currentQty": 30},
    {"id": "P007", "name": "Minyak Goreng Bimoli 1L", "barcode": "8991102221015", "category": "Bahan Pokok", "price": 18000, "minStock": 5, "capacity": 30, "currentQty": 8},
    {"id": "P008", "name": "Gula Pasir 1kg", "barcode": "8991102331012", "category": "Bahan Pokok", "price": 15000, "minStock": 3, "capacity": 25, "currentQty": 15},
]

products_df = pd.DataFrame(products)
print("DAFTAR PRODUK")
print("=" * 70)
products_df[['id', 'name', 'category', 'price', 'capacity', 'currentQty', 'minStock']]
```

```python
# Generate 60 hari history transaksi realistis
np.random.seed(42)
base_date = datetime.now() - timedelta(days=60)
weather_options = ['Cerah', 'Hujan', 'Mendung']

def generate_daily_data(product, days=60):
    """Generate daily stock data with features for ML model."""
    records = []
    qty = product['currentQty'] + np.random.randint(20, 50)

    for day in range(days):
        date = base_date + timedelta(days=day)
        day_of_week = date.weekday()
        is_weekend = 1 if day_of_week >= 5 else 0
        weather = np.random.choice(weather_options, p=[0.5, 0.3, 0.2])

        # Sales pattern: more on weekdays, affected by weather
        base_sales = np.random.poisson(3 if not is_weekend else 2)
        if weather == 'Hujan':
            base_sales = max(0, base_sales - 1)
        elif weather == 'Cerah':
            base_sales += np.random.randint(0, 2)

        actual_sales = min(base_sales, qty)
        qty -= actual_sales

        # Restock logic
        restock = 0
        if qty < product['minStock'] * 2 and np.random.random() < 0.3:
            restock = np.random.randint(10, 30)
            qty += restock

        records.append({
            'Item_ID': product['id'],
            'Item_Name': product['name'],
            'Category': product['category'],
            'Date': date,
            'Day_of_Week': day_of_week,
            'Is_Weekend': is_weekend,
            'Weather': weather,
            'Capacity': product['capacity'],
            'Amt_Available': qty,
            'Sales': actual_sales,
            'Restock': restock,
            'Price': product['price'],
        })

    return records, qty

all_records = []
for product in products:
    recs, final_qty = generate_daily_data(product)
    all_records.extend(recs)
    product['currentQty'] = final_qty

df = pd.DataFrame(all_records)
df = df.sort_values(['Date', 'Item_ID']).reset_index(drop=True)

print(f"Dataset: {len(df)} records ({len(products)} produk x 60 hari)")
print(f"Kolom: {list(df.columns)}")
print(f"\nWeather distribution:\n{df['Weather'].value_counts().to_dict()}")
print(f"\nSample data:")
df.head(10)
```

## 3. Exploratory Data Analysis

```python
# Statistik per produk
summary = df.groupby('Item_Name').agg(
    Total_Sales=('Sales', 'sum'),
    Avg_Daily_Sales=('Sales', 'mean'),
    Std_Sales=('Sales', 'std'),
    Min_Stock=('Amt_Available', 'min'),
    Max_Stock=('Amt_Available', 'max'),
    Current_Stock=('Amt_Available', 'last'),
).round(2)

print("STATISTIK PER PRODUK")
print("=" * 70)
summary
```

```python
# Visualisasi stok harian per produk
fig, axes = plt.subplots(4, 2, figsize=(14, 16))
axes = axes.flatten()

for idx, product in enumerate(products):
    ax = axes[idx]
    prod_data = df[df['Item_ID'] == product['id']]
    ax.plot(prod_data['Date'], prod_data['Amt_Available'], 'b-', linewidth=1.2)
    ax.axhline(y=product['minStock'], color='red', linestyle='--', alpha=0.7, label=f"Min Stock ({product['minStock']})")
    ax.set_title(product['name'], fontsize=10, fontweight='bold')
    ax.set_ylabel('Stok')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.suptitle('Stok Harian Per Produk (60 Hari)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. Feature Engineering & Preprocessing

```python
# Feature engineering
df['Rolling_Sales_7d'] = df.groupby('Item_ID')['Sales'].transform(lambda x: x.rolling(7, min_periods=1).mean())
df['Rolling_Sales_3d'] = df.groupby('Item_ID')['Sales'].transform(lambda x: x.rolling(3, min_periods=1).mean())
df['Cumulative_Sales'] = df.groupby('Item_ID')['Sales'].cumsum()
df['Stock_Ratio'] = df['Amt_Available'] / df['Capacity']
df['Days_Since_Start'] = (df['Date'] - df['Date'].min()).dt.days

# Target variable: Predicted Sales (next day)
df['Target_Sales'] = df.groupby('Item_ID')['Sales'].shift(-1)
df = df.dropna(subset=['Target_Sales'])

print(f"Dataset setelah feature engineering: {len(df)} records")
print(f"Features: {df.columns.tolist()}")
df.head()
```

```python
# One-hot encoding
categorical_cols = ['Category', 'Weather']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Select features
feature_cols = [c for c in df_encoded.columns if c not in [
    'Item_ID', 'Item_Name', 'Date', 'Target_Sales'
]]

X = df_encoded[feature_cols].values
y = df_encoded['Target_Sales'].values

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nFeature columns ({len(feature_cols)}):")
for i, col in enumerate(feature_cols):
    print(f"  {i+1}. {col}")
```

```python
# MinMaxScaler
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).ravel()

# Train/test split (kronologis - 80/20)
split_idx = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled[:split_idx], X_scaled[split_idx:]
y_train, y_test = y_scaled[:split_idx], y_scaled[split_idx:]

print(f"Train set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
print(f"Train ratio: {X_train.shape[0] / len(X_scaled):.1%}")
```

## 5. Model Training

### 5a. Linear Regression (Baseline)

```python
# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict
y_pred_lr = lr_model.predict(X_test)

# Inverse transform
y_test_actual = scaler_y.inverse_transform(y_test.reshape(-1, 1)).ravel()
y_pred_lr_actual = scaler_y.inverse_transform(y_pred_lr.reshape(-1, 1)).ravel()
y_pred_lr_actual = np.maximum(y_pred_lr_actual, 0)  # clamp

# Metrics
mae_lr = mean_absolute_error(y_test_actual, y_pred_lr_actual)
rmse_lr = np.sqrt(mean_squared_error(y_test_actual, y_pred_lr_actual))
r2_lr = r2_score(y_test_actual, y_pred_lr_actual)

print("LINEAR REGRESSION - Metrics")
print("=" * 40)
print(f"MAE  : {mae_lr:.4f}")
print(f"RMSE : {rmse_lr:.4f}")
print(f"R²   : {r2_lr:.4f}")
```

### 5b. Gradient Boosting Regressor

```python
# Gradient Boosting Regressor (referensi: parita2003)
gbr_model = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    subsample=0.8,
    random_state=42
)
gbr_model.fit(X_train, y_train)

# Predict
y_pred_gbr = gbr_model.predict(X_test)

# Inverse transform
y_pred_gbr_actual = scaler_y.inverse_transform(y_pred_gbr.reshape(-1, 1)).ravel()
y_pred_gbr_actual = np.maximum(y_pred_gbr_actual, 0)

# Metrics
mae_gbr = mean_absolute_error(y_test_actual, y_pred_gbr_actual)
rmse_gbr = np.sqrt(mean_squared_error(y_test_actual, y_pred_gbr_actual))
r2_gbr = r2_score(y_test_actual, y_pred_gbr_actual)

print("GRADIENT BOOSTING REGRESSOR - Metrics")
print("=" * 40)
print(f"MAE  : {mae_gbr:.4f}")
print(f"RMSE : {rmse_gbr:.4f}")
print(f"R²   : {r2_gbr:.4f}")
```

## 6. Perbandingan Model

```python
# Tabel perbandingan
comparison = pd.DataFrame({
    'Model': ['Linear Regression', 'Gradient Boosting'],
    'MAE': [mae_lr, mae_gbr],
    'RMSE': [rmse_lr, rmse_gbr],
    'R²': [r2_lr, r2_gbr],
})

# Accuracy-like metric (within 5% threshold) - referensi parita2003
threshold = 0.05
within_lr = np.sum(np.abs(y_test - y_pred_lr) <= threshold) / len(y_test)
within_gbr = np.sum(np.abs(y_test - y_pred_gbr) <= threshold) / len(y_test)
comparison['Accuracy (5% threshold)'] = [within_lr, within_gbr]

print("PERBANDINGAN MODEL")
print("=" * 60)
comparison
```

### Overfitting Check - Cross Validation


```python
# Cross-validation untuk cek overfitting
from sklearn.model_selection import cross_val_score, TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)

# Linear Regression CV
cv_lr = cross_val_score(LinearRegression(), X_scaled, y_scaled, cv=tscv, scoring='r2')
print("LINEAR REGRESSION - Cross Validation (TimeSeriesSplit, 5 folds)")
print(f"  R² per fold: {[f'{s:.3f}' for s in cv_lr]}")
print(f"  Mean R²: {cv_lr.mean():.3f} ± {cv_lr.std():.3f}")
print()

# Gradient Boosting CV
cv_gbr = cross_val_score(gbr_model, X_scaled, y_scaled, cv=tscv, scoring='r2')
print("GRADIENT BOOSTING - Cross Validation (TimeSeriesSplit, 5 folds)")
print(f"  R² per fold: {[f'{s:.3f}' for s in cv_gbr]}")
print(f"  Mean R²: {cv_gbr.mean():.3f} ± {cv_gbr.std():.3f}")
print()

# Overfitting check
train_r2_lr = r2_score(y_train, lr_model.predict(X_train))
train_r2_gbr = r2_score(y_train, gbr_model.predict(X_train))

print("OVERFITTING CHECK (Train R² vs Test R²)")
print("=" * 50)
print(f"  {'Model':<25} {'Train R²':>10} {'Test R²':>10} {'Gap':>8}")
print(f"  {'-'*25} {'-'*10} {'-'*10} {'-'*8}")
print(f"  {'Linear Regression':<25} {train_r2_lr:>10.3f} {r2_lr:>10.3f} {train_r2_lr - r2_lr:>8.3f}")
print(f"  {'Gradient Boosting':<25} {train_r2_gbr:>10.3f} {r2_gbr:>10.3f} {train_r2_gbr - r2_gbr:>8.3f}")
print()
print("Gap > 0.1 = kemungkinan overfitting")
print("Gap > 0.3 = overfitting signifikan")
```

## 7. Visualisasi - Actual vs Predicted

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Linear Regression - Scatter
ax = axes[0, 0]
ax.scatter(y_test_actual, y_pred_lr_actual, alpha=0.5, s=20, c='blue')
ax.plot([0, max(y_test_actual)], [0, max(y_test_actual)], 'r--', label='Perfect Prediction')
ax.set_xlabel('Actual Sales')
ax.set_ylabel('Predicted Sales')
ax.set_title(f'Linear Regression (R²={r2_lr:.3f})')
ax.legend()
ax.grid(True, alpha=0.3)

# Gradient Boosting - Scatter
ax = axes[0, 1]
ax.scatter(y_test_actual, y_pred_gbr_actual, alpha=0.5, s=20, c='green')
ax.plot([0, max(y_test_actual)], [0, max(y_test_actual)], 'r--', label='Perfect Prediction')
ax.set_xlabel('Actual Sales')
ax.set_ylabel('Predicted Sales')
ax.set_title(f'Gradient Boosting (R²={r2_gbr:.3f})')
ax.legend()
ax.grid(True, alpha=0.3)

# Linear Regression - Residuals
ax = axes[1, 0]
residuals_lr = y_test_actual - y_pred_lr_actual
ax.scatter(y_pred_lr_actual, residuals_lr, alpha=0.5, s=20, c='blue')
ax.axhline(y=0, color='r', linestyle='--')
ax.set_xlabel('Predicted')
ax.set_ylabel('Residuals')
ax.set_title('Residuals - Linear Regression')
ax.grid(True, alpha=0.3)

# Gradient Boosting - Residuals
ax = axes[1, 1]
residuals_gbr = y_test_actual - y_pred_gbr_actual
ax.scatter(y_pred_gbr_actual, residuals_gbr, alpha=0.5, s=20, c='green')
ax.axhline(y=0, color='r', linestyle='--')
ax.set_xlabel('Predicted')
ax.set_ylabel('Residuals')
ax.set_title('Residuals - Gradient Boosting')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Q-Q Plot residuals
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

stats.probplot(residuals_lr, dist="norm", plot=axes[0])
axes[0].set_title('Q-Q Plot - Linear Regression')
axes[0].grid(True, alpha=0.3)

stats.probplot(residuals_gbr, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot - Gradient Boosting')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Histogram residuals
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(residuals_lr, bins=30, edgecolor='k', alpha=0.7, color='blue')
axes[0].set_title('Histogram Residuals - Linear Regression')
axes[0].set_xlabel('Residual')
axes[0].set_ylabel('Frequency')

axes[1].hist(residuals_gbr, bins=30, edgecolor='k', alpha=0.7, color='green')
axes[1].set_title('Histogram Residuals - Gradient Boosting')
axes[1].set_xlabel('Residual')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Training Loss Curve (Gradient Boosting)
train_loss = [mean_squared_error(y_train, y_pred)
              for y_pred in gbr_model.staged_predict(X_train)]
val_loss = [mean_squared_error(y_test, y_pred)
            for y_pred in gbr_model.staged_predict(X_test)]

plt.figure(figsize=(10, 5))
plt.plot(train_loss, label='Training Loss', color='blue')
plt.plot(val_loss, label='Validation Loss', color='red')
plt.xlabel('Boosting Iterations')
plt.ylabel('Mean Squared Error')
plt.title('Gradient Boosting - Training vs Validation Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Accuracy over thresholds (referensi parita2003)
thresholds = np.linspace(0, 0.5, 50)

acc_lr = [np.sum(np.abs(y_test - y_pred_lr) <= t) / len(y_test) for t in thresholds]
acc_gbr = [np.sum(np.abs(y_test - y_pred_gbr) <= t) / len(y_test) for t in thresholds]

plt.figure(figsize=(10, 5))
plt.plot(thresholds * 100, acc_lr, 'b-', label='Linear Regression', linewidth=2)
plt.plot(thresholds * 100, acc_gbr, 'g-', label='Gradient Boosting', linewidth=2)
plt.axvline(x=5, color='r', linestyle='--', alpha=0.7, label='5% Threshold')
plt.xlabel('Threshold (%)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Threshold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 9. Forecast 14 Hari & Stockout Prediction

```python
# Forecast menggunakan model terbaik (Gradient Boosting)
print("FORECAST 14 HARI KE DEPAN (Gradient Boosting)")
print("=" * 70)

for product in products:
    prod_data = df[df['Item_ID'] == product['id']].copy()
    if prod_data.empty:
        continue

    # Simple forecast: gunakan rata-rata prediksi sales untuk estimasi stok
    prod_encoded = df_encoded[df_encoded['Item_ID'] == product['id']]
    if len(prod_encoded) == 0:
        continue

    last_row = prod_encoded[feature_cols].iloc[-1:].values
    last_row_scaled = scaler_X.transform(last_row)
    predicted_daily_sales = scaler_y.inverse_transform(
        gbr_model.predict(last_row_scaled).reshape(-1, 1)
    ).ravel()[0]
    predicted_daily_sales = max(0, predicted_daily_sales)

    # Forecast stok
    current_stock = product['currentQty']
    print(f"\n  {product['name']} (stok: {current_stock}, prediksi sales: {predicted_daily_sales:.1f}/hari)")
    print(f"  {'Hari':<6} {'Tanggal':<12} {'Stok':>6} {'Status':<10}")
    print(f"  {'-'*6} {'-'*12} {'-'*6} {'-'*10}")

    stockout_day = None
    for day in range(1, 15):
        future_date = datetime.now() + timedelta(days=day)
        future_stock = max(0, current_stock - predicted_daily_sales * day)
        status = 'HABIS' if future_stock <= 0 else ('RENDAH' if future_stock <= product['minStock'] else 'Aman')
        print(f"  {day:<6} {future_date.strftime('%Y-%m-%d'):<12} {future_stock:>6.1f} {status:<10}")
        if future_stock <= 0 and stockout_day is None:
            stockout_day = day

    if stockout_day:
        print(f"  >>> STOCKOUT dalam {stockout_day} hari!")
```

## 10. Ringkasan Risiko

```python
risk_data = []
for product in products:
    prod_encoded = df_encoded[df_encoded['Item_ID'] == product['id']]
    if len(prod_encoded) == 0:
        continue

    last_row = prod_encoded[feature_cols].iloc[-1:].values
    last_row_scaled = scaler_X.transform(last_row)
    pred_sales = scaler_y.inverse_transform(
        gbr_model.predict(last_row_scaled).reshape(-1, 1)
    ).ravel()[0]
    pred_sales = max(0, pred_sales)

    days_to_stockout = product['currentQty'] / pred_sales if pred_sales > 0 else None

    risk = 'Aman'
    if days_to_stockout and days_to_stockout <= 7:
        risk = 'KRITIS'
    elif days_to_stockout and days_to_stockout <= 14:
        risk = 'WASPADA'
    elif product['currentQty'] <= product['minStock']:
        risk = 'RENDAH'

    risk_data.append({
        'Produk': product['name'],
        'Stok': product['currentQty'],
        'Min Stock': product['minStock'],
        'Pred Sales/hari': round(pred_sales, 2),
        'Hari ke Stockout': round(days_to_stockout, 1) if days_to_stockout else '-',
        'Risiko': risk,
    })

risk_df = pd.DataFrame(risk_data)
risk_df['_sort'] = pd.to_numeric(risk_df['Hari ke Stockout'], errors='coerce')
risk_df = risk_df.sort_values('_sort', na_position='last').drop(columns='_sort')
print("RINGKASAN RISIKO STOCKOUT")
print("=" * 70)
risk_df
```

## 11. Kesimpulan

### Perbandingan Model:
| Aspek | Linear Regression | Gradient Boosting |
|-------|------------------|-------------------|
| Kompleksitas | Rendah | Tinggi |
| Interpretability | Tinggi (slope jelas) | Rendah (black box) |
| Akurasi | Baseline | Lebih tinggi |
| Training time | Cepat | Lebih lama |
| Overfitting risk | Rendah | Perlu tuning |

### Rekomendasi:
- **Linear Regression** cocok untuk quick insight dan interpretasi tren (dipakai di website StokManager)
- **Gradient Boosting** cocok untuk prediksi akurat dengan banyak fitur (weather, day_of_week, rolling avg)
- Untuk production di website, Linear Regression sudah cukup karena ringan dan bisa jalan di browser
- Gradient Boosting bisa dipakai sebagai backend service jika butuh akurasi lebih tinggi

### Fitur yang meningkatkan akurasi:
1. `Rolling_Sales_7d` — moving average 7 hari
2. `Day_of_Week` / `Is_Weekend` — pola mingguan
3. `Weather` — pengaruh cuaca terhadap penjualan
4. `Stock_Ratio` — rasio stok terhadap kapasitas
